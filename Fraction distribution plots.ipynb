```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Функция для отображения гистограммы явки

```python
def draw_turnout_distribution(sizes, turnout, bins, title):
    sizes_in_people = np.floor(sizes)
    turnout_in_people = np.floor(sizes_in_people * turnout)
    turnout_fraction = turnout_in_people / sizes_in_people

    plt.figure(figsize=(16,8))
    plt.hist(turnout_fraction, bins=bins, range=(0, 1.0))
    plt.xticks([0.0, 1.0, 1./2, 1./3, 2./3, 1./4, 3./4, 2./5, 3./5, 2./7, 3./7, 4./7, 5./7, 3./8, 5./8,
                4./9, 5./9, 3./10, 7./10],
               ["0", "1", "1/2", "1/3", "2/3", "1/4", "3/4", "2/5", "3/5", "2/7", "3/7", "4/7", "5/7", "3/8", "5/8",
                "4/9", "5/9", "3/10", "7/10"],
               rotation=90)
    plt.xlim((0, 1.0))
    plt.xlabel("Явка")
    plt.ylabel("Число УИКов с такой явкой")
    plt.title("Распределение явки по числу УИКов, %s" % title);
```

## Гистограмма явки для искусственного случая

Всего 50.000 участков. Размеры участков распределены равномерно от 1000 до 2000. Явка распределена нормально $\mu = 0.55$, $\sigma = 0.1$.

```python
N_POINTS = 50000
sizes = np.random.uniform(1000, 2000, N_POINTS)
turnout = np.random.normal(0.55, 0.1, N_POINTS)

draw_turnout_distribution(sizes, turnout, 1000, "участки uniform(1000, 2000), явка norm(0.55, 0.1)")
```

## Гистограмма явки для случая, приближенного к натуральному

Всего 95.000 участков. Размеры участков распределены по гамме $\theta = 2.5$, $k=500$ ($E = 1250$). Явка распределена по бете $\alpha = 8$, $\beta = 4$ ($E = 0.67$).

```python
N_POINTS = 95000
sizes = np.random.gamma(2.5, 500, N_POINTS)
turnout = np.random.beta(8, 4, N_POINTS)

draw_turnout_distribution(sizes, turnout, 1000, "участки gamma(2.5, 500), явка beta(8, 4)")
```
